```python
def chunks(lst, n):
    """Yield successive n-sized chunks."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
batch_size = 100
for id_chunk, emb_chunk, text_chunk in zip(chunks(ids, batch_size), chunks(embeddings.tolist(), batch_size), chunks(clean_texts, batch_size)):
    vectors = [
        {"id": id_, "values": vector, "metadata":{"text":text}}
        for id_, vector, text in zip(id_chunk, emb_chunk, text_chunk)
    ]
    index.upsert(vectors=vectors)
print(vectors)
```
